## モデル
$$
X_t \mid \eta_0,\alpha,\mathbf y
\sim
\text{Poisson}\!\left(\exp(\eta_0+\alpha y_t)\right)
$$

$$
\mathbf y \sim \mathcal N(\mathbf 0, K^{(\mathrm{exp})})
$$

$$
K^{(\mathrm{exp})}_{ij}
=
\exp\!\left(-\frac{|t_i-t_j|}{l}\right)
$$

$$
\theta = \exp(-1/l), \qquad
l = -\frac{1}{\log \theta}
$$

$$
\eta_0 \sim \mathcal N(\eta_0^{init}, c\sigma_{\eta_0}^2)
$$

$$
\alpha \sim \mathcal N^+(\alpha^{init}, c\sigma_{\alpha}^2)
$$

$$
\log l
\sim
\mathcal N\!\left(
\log\!\left(-\frac{1}{\log \theta^{init}}\right),
c\sigma_{l}^2
\right)
$$

```python
from src.models_poisson_exp import sample_exp_kernel_model

trace_exp = sample_exp_kernel_model(
    data=df["ALL_norm"],
    lambda0_init=lambda0_init,
    alpha_init=alpha_init,
    theta_init=theta_init,
    lambda0_se=lambda0_se,
    alpha_se=alpha_se,
    theta_se=theta_se,
    y_init=y_init,
    prior_scale=1.0,
    draws=2000,
    tune=2000,
    chains=4,
    target_accept=0.95,
)
```

```python
import arviz as az

az.summary(trace_exp, var_names=["lambda0" , "alpha" , "theta"])
```

```python
az.plot_trace(trace_exp, var_names=["lambda0", "alpha", "theta"])
plt.tight_layout()
plt.show()
```
